```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import metrics
```

```python
def funkcja1(x):
    return np.sin(x)

def funkcja2(x):
    return np.sin(np.power(x, -1))

def funkcja3(x):
    return np.sign(np.sin(8*x))
```

```python
x = np.linspace(1e-6, 10, 10000)  #(początkowa wartość, końcowa wartość, liczba próbek)
plt.figure(figsize=[12,8])
plt.plot(x, funkcja1(x))
plt.plot(x, funkcja2(x))
plt.plot(x, funkcja3(x))
plt.plot(x, funkcja1(x), label='sin(x)')
plt.plot(x, funkcja1(x), label='sin(x^(-1))')
plt.plot(x, funkcja3(x), label='sgn(sin(8x))')
plt.legend()
plt.show()
```

```python
n_probek = 10
n_oczekiwana = 10000
x = np.linspace(1e-6, 2*np.pi, n_probek)
y=funkcja1(x)
plt.plot(x, funkcja1(x))
plt.scatter(x, funkcja1(x))
plt.plot(x, funkcja1(x), label='interpolacja sin(x)')
plt.title('Wykres sin(x) przy 10 próbkach')

x_interp = np.linspace(1e-6, 2*np.pi, n_oczekiwana)
y_interp = np.interp(x_interp, x, y)
y_true = funkcja1(x_interp)

plt.plot(x_interp, y_true, label='sin(x)')

plt.legend()
plt.show()
print(f"MSE: {metrics.mean_squared_error(y_pred=y_interp, y_true=y_true):.8f}")
```
