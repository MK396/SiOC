# Interpolacja 1D

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import metrics
from scipy import interpolate
from sklearn import metrics
```

```python
def funkcja1(x):
    return np.sin(x)

def funkcja2(x):
    return np.sin(np.power(x, -1))

def funkcja3(x):
    return np.sign(np.sin(8*x))
```

```python
x = np.linspace(1e-6, 10, 10000)  #(początkowa wartość, końcowa wartość, liczba próbek)
plt.figure(figsize=[12,8])
plt.plot(x, funkcja1(x))
plt.plot(x, funkcja2(x))
plt.plot(x, funkcja3(x))
plt.plot(x, funkcja1(x), label='sin(x)')
plt.plot(x, funkcja1(x), label='sin(x^(-1))')
plt.plot(x, funkcja3(x), label='sgn(sin(8x))')
plt.legend()
plt.show()
```

###**Interpolacja linowa**

```python
n_probek = 10
n_oczekiwana = 10000
x = np.linspace(1e-6, 2*np.pi, n_probek)
y=funkcja1(x)
plt.plot(x, funkcja1(x))
plt.scatter(x, funkcja1(x))
plt.plot(x, funkcja1(x), label='interpolacja sin(x)')
plt.title('Wykres sin(x) przy 10 próbkach')

x_interp = np.linspace(1e-6, 2*np.pi, n_oczekiwana)
y_interp = np.interp(x_interp, x, y)
y_true = funkcja1(x_interp)

plt.plot(x_interp, y_true, label='sin(x)')

plt.legend()
plt.show()
print(f"MSE: {metrics.mean_squared_error(y_pred=y_interp, y_true=y_true):.8f}")
```

# Interpolacja 2D

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
# Funkcja do zmniejszania obrazu za pomocą splotu z jądrem uśredniającym
def zmniejsz_obraz(image, wspolczynnik_skalowania):
    kernel = np.ones((wspolczynnik_skalowania, wspolczynnik_skalowania), np.float32) / (wspolczynnik_skalowania ** 2)
    wynik = cv2.filter2D(image, -1, kernel)
    return wynik[::wspolczynnik_skalowania, ::wspolczynnik_skalowania]

# Funkcja do powiększania obrazu za pomocą interpolacji najbliższego sąsiada
def powieksz_najblizszy_sasiad(image, wspolczynnik_skalowania):
    wysokosc, szerokosc = image.shape
    nowa_wysokosc, nowa_szerokosc = wysokosc * wspolczynnik_skalowania, szerokosc * wspolczynnik_skalowania
    wynik = cv2.resize(image, (nowa_szerokosc, nowa_wysokosc), interpolation=cv2.INTER_NEAREST)
    return wynik

# Funkcja do powiększania obrazu za pomocą interpolacji dwuliniowej
def powieksz_dwuliniowa(image, wspolczynnik_skalowania):
    wysokosc, szerokosc = image.shape
    nowa_wysokosc, nowa_szerokosc = wysokosc * wspolczynnik_skalowania, szerokosc * wspolczynnik_skalowania
    wynik = cv2.resize(image, (nowa_szerokosc, nowa_wysokosc), interpolation=cv2.INTER_LINEAR)
    return wynik
```

```python
# Oryginalny obraz
oryginalny_obraz = cv2.imread("/content/drive/MyDrive/pies1.jpg", cv2.IMREAD_GRAYSCALE)
wspolczynnik_skalowania = 2

# Zmniejszanie obrazu
zmniejszony_obraz = zmniejsz_obraz(oryginalny_obraz, wspolczynnik_skalowania)

# Powiększanie obrazu - Interpolacja najbliższego sąsiada
powiekszony_obraz_najblizszy_sasiad = powieksz_najblizszy_sasiad(zmniejszony_obraz, wspolczynnik_skalowania)

# Powiększanie obrazu - Interpolacja dwuliniowa
powiekszony_obraz_dwuliniowa = powieksz_dwuliniowa(zmniejszony_obraz, wspolczynnik_skalowania)
```

```python
cv2_imshow(oryginalny_obraz)
cv2_imshow(zmniejszony_obraz)
cv2_imshow(powiekszony_obraz_najblizszy_sasiad)
print(f"MSE (Najbliższy sąsiad): {metrics.mean_squared_error(oryginalny_obraz, powiekszony_obraz_najblizszy_sasiad):.8f}")
cv2_imshow(powiekszony_obraz_dwuliniowa)
print(f"MSE (Dwuliniowa): {metrics.mean_squared_error(oryginalny_obraz, powiekszony_obraz_dwuliniowa):.8f}")
```
